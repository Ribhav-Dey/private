```python
import pandas as pd

msft = yf.Ticker('CRM')
news_data = msft.news
extracted_data = []
for news_item in news_data:
    publisher = news_item.get('publisher')
    link = news_item.get('link')
    provider_publish_time = news_item.get('providerPublishTime')
    extracted_data.append({'publisher': publisher, 'link': link, 'providerPublishTime': provider_publish_time})

# Create a DataFrame
news_df = pd.DataFrame(extracted_data)
```

```python
print(news_df)
```

```python
print(news_df.iloc[0]['link'])
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_news_content(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Adjust the scraping logic based on the website structure
    article = soup.find('div', class_='caas-body')
    if article is None:
        article = soup.find('article')
    if article is None:
        article = soup.find('div', {'class': 'content'})

    if article:
        return article.text.strip()
    else:
        return "Content not found"
    return article


# Scrape the content for each news item
for item in extracted_data:
    item['content'] = scrape_news_content(item['link'])

# Create a DataFrame
news_df = pd.DataFrame(extracted_data)
```

```python
# Remove rows where content is "Content not found"
news_df = news_df[news_df['content'] != "Content not found"]

# Save the DataFrame to a CSV file
news_df.to_csv('msft_news.csv', index=False)

print(news_df)
```

```python
# Convert the news DataFrame to a list of document dictionaries
news_document = news_df[['content']]
##news_documents = [{'id_': i, 'text': row['content']} for i, row in news_df.iterrows()]
news_document
```

```python
#!pip install fpdf
#!pip install pyPDF2
import pandas as pd
from fpdf import FPDF
from PyPDF2 import PdfReader, PdfWriter

# Sample data assuming you have a dataframe named news_document
# news_document = pd.read_csv('your_file.csv')  # Example of loading the dataframe

# Function to clean text
def clean_text(text):
    return text.encode('latin-1', 'ignore').decode('latin-1')

# Apply the cleaning function to the 'content' column
news_document['content'] = news_document['content'].apply(clean_text)

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'News Content', 0, 1, 'C')

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Page {self.page_no()}', 0, 0, 'C')

    def chapter_body(self, body):
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, body)
        self.ln()

# Create instance of FPDF
pdf = PDF()
pdf.add_page()

# Add content from dataframe to PDF
for content in news_document['content']:
    pdf.chapter_body(content)

# Save the new content PDF to a temporary file
new_pdf_path = 'latest_content.pdf'
pdf.output(new_pdf_path)

# Append the new content to the existing PDF
existing_pdf_path = 'news_content.pdf'
output_pdf_path = 'news_content.pdf'

# Read the existing PDF
reader = PdfReader(existing_pdf_path)
writer = PdfWriter()

# Add all pages from the existing PDF to the writer
for page in reader.pages:
    writer.add_page(page)

# Read the new content PDF
new_reader = PdfReader(new_pdf_path)

# Add all pages from the new content PDF to the writer
for page in new_reader.pages:
    writer.add_page(page)

# Write the combined PDF to a file
with open(output_pdf_path, 'wb') as output_pdf:
    writer.write(output_pdf)

print(f"Combined PDF saved as {output_pdf_path}")
```
